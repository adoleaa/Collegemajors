Visualizing earnings based on College majors

```
import pandas as pd # importing libraries required
import matplotlib.pyplot as plt
%matplotlib inline
```

```
recent_grads = pd.read_csv("https://raw.githubusercontent.com/adoleaa/data/master/college-majors/grad-students.csv") #reading the datasets into the data frame
```

```
print(recent_grads.iloc[0])
print(recent_grads.head())
print(recent_grads.tail())
recent_grads = recent_grads.dropna()
```

```
recent_grads.plot(x='Grad_sample_size', y="Grad_median", kind='scatter')  # scatter plots with Pandas
```

```
recent_grads.plot(x="Grad_sample_size", y="Grad_unemployment_rate", kind="scatter")
```

```
recent_grads.plot(x="Grad_full_time_year_round", y="Grad_median", kind="scatter")
```

```
recent_grads.plot(x="Grad_share", y="Grad_unemployment_rate", kind="scatter")
```

```
recent_grads.plot(x="Grad_sample_size", y="Grad_unemployed", kind="scatter")
```

```
recent_grads.plot(x="Grad_sample_size", y="Grad_employed", kind="scatter")
```

```
recent_grads.plot(x='Grad_employed', y='Grad_median', kind='scatter')
```

```
recent_grads.plot(x='Grad_unemployed', y='Grad_median', kind='scatter')
```

```
cols =["Grad_employed", "Grad_unemployed", "Grad_unemployment_rate", "Grad_unemployment_rate", "Nongrad_employed", "Nongrad_unemployed" ] # Pandas Histograms

fig = plt.figure(figsize=(5,12))
for r in range(0,4):
   ax = fig.add_subplot(4,1,r+1)
   ax = recent_grads[cols[r]].plot(kind="hist",rot=30)
```

```
cols =["Grad_employed", "Grad_unemployed", "Grad_unemployment_rate", "Grad_unemployment_rate", "Nongrad_employed", "Nongrad_unemployed" ] # Pandas Histograms

fig = plt.figure(figsize=(5,12))
for r in range(4,6):
   ax = fig.add_subplot(4,1,r-1)
   ax = recent_grads[cols[r]].plot(kind="hist",rot=30)
```

```
from pandas.plotting import scatter_matrix
scatter_matrix(recent_grads[['Grad_sample_size', 'Grad_median']], figsize=(6,6)) # Pandas, Scatter Matrix Plot
```
